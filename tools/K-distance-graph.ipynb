```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()
# 最值归一化处理
scaler = MinMaxScaler()
data = scaler.fit_transform(iris.data)
distance_matrix = squareform(pdist(data))
k=4
KNearestNeighbor = np.sort(distance_matrix)[:,k]
KNearestNeighbor = np.sort(KNearestNeighbor)
KNearestNeighbor
X=range(len(KNearestNeighbor))
Y=KNearestNeighbor
plt.plot(X,Y)
#plt.xticks(X)
plt.grid()
plt.show()
```
